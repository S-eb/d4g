```python
import polars as pl
import json
```

```python
# Opening JSON file
f = open('data.json')
```

```python
# returns JSON object as a dictionary
data = json.load(f)
```

```python

out_json = {"list":[]}
for kd in data["data"]:
    for k,v in kd.items():
        sub_kd = {}
        sub_kd["code"] = k
        sub_kd["value"] = v
        out_json["list"].append(sub_kd)
```

```python
df = pl.DataFrame(out_json['list']).unnest('value').unnest('res')
```
